```python
A = pd.read_excel('county-to-county-2013-2017-ins-outs-nets-gross.xlsx', sheet_name="Wyoming")
As = A.drop(A.index[[0,1]])
SCAl = As.columns[0]
CCAl = As.columns[1]
SCBl = As.columns[2]
CCBl = As.columns[3]
B2Al = As.columns[8]
B2AMOEl = As.columns[9]
A2Bl = As.columns[10]
A2BMOEl = As.columns[11]
LL = [(i,As[SCAl].tolist()[i], As[CCAl].tolist()[i], As[SCBl].tolist()[i], As[CCBl].tolist()[i]) for i in range(len(As[SCAl].tolist()))]
rlist = []
tempSCA = []
tempCCA = []
tempSCB = []
tempCCB = []
for i,SCA,CCA,SCB,CCB in LL:
    if np.isnan(np.float64(CCB)):
        rlist.append(i)
        SCB = 'NAN'
        CCB = 'NAN'
    else:
        SCB = str(np.int32(SCB))
        CCB = str(np.int32(CCB))
    
    if len(SCB) == 0:
        rlist.append(i)
        SCB = 'NAN'
        
    elif len(SCB) == 1:
        SCB = '00' + SCB
    
    elif len(SCB) == 2:
        SCB = '0' + SCB
    
    if len(CCB) == 0:
        rlist.append(i)
        CCB = 'NAN'
        
    elif len(CCB) == 1:
        CCB = '00' + CCB
    
    elif len(CCB) == 2:
        CCB = '0' + CCB
        
    if (SCB[0] is not '0'):
        rlist.append(i)
        SCB = 'NAN'
        CCB = 'NAN'
        
    if (SCA[0] is not '0'):
        rlist.append(i)
        SCA = 'NAN'
        CCA = 'NAN'
        
    if SCB == '072':
        rlist.append(i)
    
    tempSCA.append(SCA)
    tempCCA.append(CCA)
        
    tempSCB.append(SCB)
    tempCCB.append(CCB)
rlist = np.unique(rlist)
C = As.drop(As.index[rlist])
fipsA = np.int32(C[SCAl] + C[CCAl])
fipsB = np.int32(C[SCBl] + C[CCBl])
fipsA[fipsA==2230]=2232
fipsB[fipsB==2230]=2232
counties = cf.convFIPS2Name(fipsA)
counties = cf.convFIPS2Name(fipsB)
dl.load_data('idx_dict.obj')
FLOW = dl.load_data('temp_FLOW.obj')
d = np.int32([fipsA, fipsB, C[B2Al], C[B2AMOEl], C[A2Bl], C[A2BMOEl]]).T
D = pd.DataFrame(data=d, columns=['fipsA', 'fipsB', 'B2A', 'B2As', 'A2B', 'A2Bs'])

```
